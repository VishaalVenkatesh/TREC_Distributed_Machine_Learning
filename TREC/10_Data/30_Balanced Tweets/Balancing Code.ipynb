```python
i = 0
for file in a:
    name = ['cyclone', 'earthquake', 'fire', 'flood', 'shooting']
    df = pd.read_csv(file)
    df_c = df[df['Priority'] == 'Critical']
    shape_c = df_c.shape
    df_not_c = df[df['Priority'] != 'Critical']
    df_not_c_bal = df_not_c.sample(shape_c[0]) 
    df_combine = pd.concat([df_c, df_not_c_bal])
    df_combine.to_csv(name[i]+'_TREC_2019_B_test_BALANCED.csv', index = False)
    i = i+1
```
